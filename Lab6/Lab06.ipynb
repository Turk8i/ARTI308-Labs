```python
# 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Load the dataset
customers = pd.read_csv('Ecommerce Customers')
```

```python
# 2
# Check the first few rows
customers.head()

# Check the data types and if there are any missing values
customers.info()

# Get a statistical summary of the numerical columns
customers.describe()
```

3

Looking at the output of info(), the Ecommerce Customers dataset is already clean and does not have missing values, so no imputation or dropping of rows is required for this step.

4

For this specific dataset, the numerical features are already well-suited for Linear Regression. We just need to separate our numerical features from the categorical ones (Email, Address, and Avatar), which we will not use for the model.

```python
#5 
# Define Features (X) and Target (y)
X = customers[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = customers['Yearly Amount Spent']

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#6
from sklearn.linear_model import LinearRegression

# Instantiate the model
lm = LinearRegression()

# Train/fit the model on the training data
lm.fit(X_train, y_train)

# (Optional) You can check the coefficients to see the weight of each feature
print(lm.coef_)
```

```python
#7
# Make predictions on the test set
predictions = lm.predict(X_test)

# Evaluate the metrics
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
